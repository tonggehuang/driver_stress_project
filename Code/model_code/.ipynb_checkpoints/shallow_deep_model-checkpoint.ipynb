```python
import os
import re
import cv2
import math

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import datetime
import warnings
warnings.filterwarnings("ignore")
```

#### EMP006 data process

```python
def percentile_label(df):
    df.loc[df.abs_gra_eda >= df.abs_gra_eda.quantile(0.75), 'stress_level'] = 1
    df.loc[(df.abs_gra_eda < df.abs_gra_eda.quantile(0.75))
           &(df.abs_gra_eda >= df.abs_gra_eda.quantile(0.5)), 'stress_level'] = 2
    df.loc[df.abs_gra_eda < df.abs_gra_eda.quantile(0.5), 'stress_level'] = 1
    return df
```

```python
def df_append(dir_path):
    df_we_label = pd.DataFrame()
    for i in os.listdir(dir_path):
        df_path = os.path.join(dir_path,i)
        df = pd.read_csv(df_path, sep=',')
        df = percentile_label(df)
        df_we_label = df_we_label.append(df, ignore_index=True)
    return df_we_label
```

```python
df_we_label = df_append('../../model_data/emp006_we')
```
